## Visualization 1 - using Python:

### I have decided to use the following dataset found at this link: https://data.ontario.ca/dataset/rabies-cases-in-ontario
### The data is reported in a .csv file, which I will load in and manipulate here.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the relevant dataset:
rabies = pd.read_csv("csv_rabies_cases_graph_en_2026.csv")
```

```python
rabies.head()
```

```python
# Set the style
sns.set_style("whitegrid")
sns.set_palette("colorblind")

# Create the figure
plt.figure(figsize=(10,6))

# Plot the lines
sns.lineplot(data = rabies,
             x = "Year",
             y = "Number of raccoon variant cases",
             marker = "o",
             linewidth = 2.5,
             label = "Racoon Variant")

sns.lineplot(data = rabies,
             x = "Year",
             y = "Number of fox variant cases",
             marker = "s",
             linewidth = 2.5,
             label ="Fox Variant")

# Add labels
plt.title("Trends in Laboratory-Confirmed Rabies in Ontario (2015-2025)",
          fontsize = 14)

plt.xlabel("Year")

plt.ylabel("Number of Cases")

plt.legend(title = "Rabies Variant")

plt.tight_layout()

plt.savefig("Viz1_rabies_trends_PYTHON.png", dpi=300, bbox_inches = "tight")
plt.show()
```
