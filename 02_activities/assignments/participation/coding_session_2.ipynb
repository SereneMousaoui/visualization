```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

Make some sample data

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

Make a line plot with both our y variables on the same axis

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2) # overlaying 2 line plots
```

Adding a legend:

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
```

Modifying our legend:

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right', 
          frameon = True, #add frame around the legend 
          fontsize = 12, #change font size 
          ncol = 2, #specify number of columns 
          shadow = True )
```

Moving the legend outside of the plot area:

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'upper left',
          bbox_to_anchor = (1,1)) # IMPORTANT: make the top left part of the legend attach to the top right of the graph.
```

Annotatins with the text() function:

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!")
```

We can further modify the text annotation by adding arguments:

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!",
        ha = 'center', # alignment of text
        color = 'red', # modify the font color
        size = 20 # modify the font size
        )
```

Positioning text annotations:

```python
# make a simple empty set of axes:
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
```

```python
# we can position our annotation with reference to data like before, the axes, or the figure
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData ) # refer to the data coordinate 1, 5 and give it a text, then use transData argument to refer the text to a specific data point
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes ) # half way along the X axis and 10% of the way up the y axis
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure ) # fraction of the figure this time, relative to the whole figure canvas
```

Annotating with arrows:

```python
# use annotate() to add arrows and text to focus on a point:
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10,95), xytext=(20,94),
            arrowprops = dict(facecolor = 'black'))
```

```python
# use the arrowprops arguent to adjust our arrow style and colour

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10,95), xytext=(30,94),
            arrowprops = dict(arrowstyle = "<->",
                              color = 'hotpink'))

# for arrowprops, Try "->", "<->", "-[", "fancy", "wedge"
```

Removing tick marks and labels. Use locator to remove ticks and labels, use formatter to remove only labels.

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# we can also define a max number of ticks we want on our axis:
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# we can set tick marks at a particular interval using MultipleLocator

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# To rotate the axis labels and horizonal alighment - use xticks()

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

plt.xticks(rotation=45, ha='right')
```

ACTIVITY - Modifying axis title fonts.
Use what we learned to modify our x axis title.

We want a serif font with indigo colour, and for the axis title to be "Shiny New X Axis!"
Use plt.xlabel().
HINT: Recall fontdict.

```python
font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict=font1)
```

Styles are a convenient way to change many aesthetic dimensions of our plots at the same time.

```python
# To view available pre-made styles, use:
plt.style.available
```

To use a style, type the following (or. variation of it based on output of previous cell) before making the figure: 
plt.style.use('fivethirtyeight')

```python
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('X Axis!')
```
